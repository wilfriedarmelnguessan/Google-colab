# GÉNÉRER DES DONNES SIMILÉ



```python
np.random.seed(42)
num_samples=10
```

```python
age=np.random.randint(18,90,size=num_samples)
age
```

```python
income=np.random.randint(20000,100000,size= num_samples)
loan_amount=np.random.randint(5000,100000,size= num_samples)
num_accounts= np.random.randint(1,10, size= num_samples)
```

#Creation des caractéristiques d'entrée


```python
x=np.column_stack((age,income,loan_amount,num_accounts)).T
x
```

```python
#fomction d'activation Relu
def relu(x):
  return np.maximum(0,x)

#fonction d'activation sigmoide
def sigmoid(x):
  return 1/(1+ np.exp(-x))
```

```python
#nombre de neurone
hidden_size= 5
input_size= x.shape[0]
output_size=1
np.random.seed(42)
w1= np.random.rand(hidden_size,input_size)
w1
b1= np.random.rand(hidden_size,1)
w2=np.random.rand(output_size,hidden_size)
b2= np.random.rand(output_size,1)
```

```python
#forward pass
z1= np.dot(w1,x,)+b1
a1=relu(z1)
a1
```

```python
z2=np.dot(w2,a1)+b2
a2=sigmoid(z2)
a2
```
